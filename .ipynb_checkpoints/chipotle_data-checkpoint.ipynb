```python
import pandas 
import seaborn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipotle_data = pandas.read_csv(url, sep = '\t')

chipotle_data["item_price"] = chipotle_data["item_price"].str.replace("$", "", regex = True).astype(float)

chipotle_data.head()
```

```python
#Q3:
count_order_dict={}

for item in list(chipotle_data["item_name"].unique()):
    count_order_dict[item] = len(chipotle_data[chipotle_data["item_name"]== item])

sorted_item_names = sorted(count_order_dict, key= count_order_dict.get, reverse=True)
#for item_name in sorted_item_names:
#    print(item_name + " Order Count: " + str(count_order_dict[item_name]))

print(sorted_item_names[0] + 's Order Count: ' + str(count_order_dict[sorted_item_names[0]]))
```

```python
#8:
unique_count_series = chipotle_data['item_name'].value_counts(ascending=False)
most_order_item = unique_count_series.index[0] # Chicken Bowl
highest_order_count = unique_count_series[most_order_item] #726

print(most_order_item + " : " + str(highest_order_count))

item_names = unique_count_series.head(5).index
item_counts = unique_count_series.head(5)

item_count_hist = seaborn.histplot(x=item_names, y=item_counts)
item_count_hist.set(ylabel = "Item Count")
item_count_hist.set_xticklabels(item_names, rotation=40, ha='right')
```

```python
#9
item_names = unique_count_series.index
item_counts = unique_count_series

colors = seaborn.color_palette('pastel')[0:5]
plt.pie(labels = item_names, x = item_counts, colors = colors, autopct="%.2f%%"
)
plt.show()
```

```python
#10
unique_price_series = chipotle_data['item_price'].value_counts(ascending=False)
most_price_item = unique_price_series.index #Gets us the prices in order of times the item was bought
highest_price_count = unique_price_series[most_price_item] #Gets us the counts of the prices of items


piechart = seaborn.scatterplot(x = most_price_item, y = highest_price_count)
plt.title("Item Price and Order Count Scatterplot")
plt.xlabel("Item Price")
plt.ylabel("Order Count")
plt.show()
```

```python
#11
item_names = unique_count_series.head(5).index
item_counts = unique_count_series.head(5)

item_count_bar = seaborn.barplot(x=item_names, y=item_counts)
item_count_bar.set(ylabel = "Item Count")
item_count_bar.set_xticklabels(item_names, rotation=40, ha='right')
```
